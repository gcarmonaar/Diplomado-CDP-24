```python
#Crear una lista de las letras utilizando espacios en blanco
'A B C D E'.split()
#lista=["A", "B", "C", "D", "E"]
```

```python
df=pd.DataFrame(np.random.randn(6,4),
                index='A B C D E F'.split(),
                columns='Uno Dos Tres Cuatro'.split())
```

```python
print(df)
```

```python
#Dimensión
df.shape
```

```python
##Información general
df.info()
```

```python
#Agregar columna que tiene la suma de las columnas Uno y Dos
df['Nueva']= df['Uno']+ df['Dos']
print(df)
```

```python
df['Nueva']= df['Uno']+ df['Dos'] + df['Tres'] + df['Cuatro']
print(df)
```

```python
#Eliminar columna
df.drop('Nueva', axis='columns')
```

```python
print(df)
```

```python
#Eliminar columna (definitivamente)
df.drop('Nueva', axis='columns', inplace=True)
print(df)
```

```python
#Otra manera de eliminar columna tipo inplace:
del df['Cuatro']
```

```python
df
```

```python
#Eliminar fila
df.drop('E', axis='index', inplace=True)
df
```
