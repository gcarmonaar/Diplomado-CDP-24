```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url="/content/drive/MyDrive/Anahuac/Diplomado en Ciencia de Datos/Python/salary.csv"
data = pd.read_csv(url)
```

```python
data.head()
```

```python
sns.histplot(x=data["age"], bins=30, kde=True,stat="probability",hue=data["salary"], palette="pastel")
```

```python
sns.boxplot(x=data["age"],color="lightgreen", showmeans=True, legend=False)
```

```python
sns.boxplot(x=data['hours-per-week'], showmeans=True, palette="pastel")
```

```python
tab_fr_sex=data["sex"].value_counts()
```

```python
type(tab_fr_sex)
```

```python
sns.barplot(tab_fr_sex, color="lightblue")
```

```python
datos={"labels": ["A", "B", "C", "D"],
      "values": [10, 35, 25, 20]}
df=pd.DataFrame(datos)
```

```python
df.head()
```

```python
expl=[0,0.1,0,0]
```

```python
sns.set_style("whitegrid")
colores=sns.color_palette("pastel")
plt.pie(df["values"], labels=df["labels"], autopct="%1.1f%%", explode=expl, colors=colores)
plt.title("Porcentaje de...")
plt.show()
```

```python
data_sex=pd.DataFrame({"sex":tab_fr_sex.index, "frequency":tab_fr_sex.values})
data_sex.head()
```

```python
sns.set_style("darkgrid")
colores=sns.color_palette("pastel")
plt.pie(data_sex["frequency"], labels=data_sex["sex"], autopct="%1.1f%%", colors=colores)
plt.title("Porcentaje de hombres y mujeres")
plt.show()
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(10,10))
plt.bar(x=data_sex["sex"], height=data_sex["frequency"], width=0.3)
for a, b in enumerate(data_sex["frequency"]):
    plt.text(a, b, b)
plt.title("Gráfico de barras de la columa Sex")
plt.show()
```
